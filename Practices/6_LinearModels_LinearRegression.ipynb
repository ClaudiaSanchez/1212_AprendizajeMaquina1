```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

# Linear Regression

```python
X = np.random.uniform(low=-10,high=10,size=100)
Y = 10*X + 5 + np.random.normal(loc=0,scale=20,size=len(X))
print(X[:5])
print(Y[:5])
```

```python
plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
XX = np.reshape(X,(len(X),1))
print(XX.shape)
```

```python
model = LinearRegression()
model.fit( XX, Y )
```

```python
print(model.coef_)
print(model.intercept_)
```

Modelo $y = mx + b$

Modelo $y = \beta_1 x + \beta_0$

$\beta_0 intercept $

$\beta_1 coef$

```python
m,b = model.coef_[0], model.intercept_
plt.figure()
plt.scatter(X,Y)
plt.plot( [-10,10], [ -10*m+b, 10*m+b ], color='red' )
plt.show()
```

```python
plt.figure()
plt.scatter(X,Y)
plt.show()

XX = np.reshape(X,(len(X),1))
print(XX.shape)
XX2 = np.concatenate((XX,XX**2),axis=1)
print(XX2.shape)

model = model = LinearRegression()
model.fit( XX2, Y )
print(model.coef_)
print(model.intercept_)
```

# Modelo cuadr√°tico

```python
X = np.random.uniform(low=-10,high=10,size=100)
Y = 30*X**2 + 10*X + 5 + np.random.normal(loc=0,scale=500,size=len(X))

plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
XX = np.reshape(X,(len(X),1))
print(XX.shape)
```

```python
XX2 = np.concatenate((XX,XX**2),axis=1)
print(XX2.shape)
```

```python
model = LinearRegression()
model.fit(XX2,Y)
```

```python
print(model.coef_)
print(model.intercept_)
```

Modelo  $ùë¶= b_1 x + b_2 x^2  + ùëè_0$

```python
b1 = model.coef_[0]
b2 = model.coef_[1]
b0 = model.intercept_
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot( X, b1*X + b2*X**2+b0,'.',color='red' )
plt.show()
```

# Linear Regression - winequality

```python
df = pandas.read_csv('../Datasets/winequality-red.csv',sep=";")
df
```

```python
X = df.values[:,:-1]
Y = df.values[:,-1]
Xtrain, Xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.30, random_state=42)
print('X',X.shape)
print('Xtrain',Xtrain.shape)
print('Xtest',Xtest.shape)
```

```python
model = LinearRegression()
model.fit( Xtrain, ytrain)
ypred = model.predict( Xtest)
print('MSE',mean_absolute_error(ytest,ypred))
```

```python
print(df.columns)
print(model.coef_)
print(model.intercept_)
```

```python
Xtrain2 = np.concatenate((Xtrain,Xtrain**2),axis=1)
Xtest2 = np.concatenate((Xtest,Xtest**2),axis=1)
model2 = LinearRegression()
model2.fit( Xtrain2, ytrain)
ypred = model2.predict( Xtest2)
print('MSE',mean_absolute_error(ytest,ypred))
```

## Normalized features

```python
for i in range(X.shape[1]):
    print(df.columns[i],'[',np.min( df.values[:,i] ), np.max( df.values[:,i] ) ,']')
```

```python
XN = X.copy()
for i in range(X.shape[1]):
    XN[:,i] = ( X[:,i]-np.mean(X[:,i]) )/np.std(X[:,i])
    print(df.columns[i],'[',np.min( XN[:,i] ), np.max( XN[:,i] ) ,']')
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(XN, Y, test_size=0.30, random_state=42)
print('X',X.shape)
print('Xtrain',Xtrain.shape)
print('Xtest',Xtest.shape)
```

```python
model = LinearRegression()
model.fit( Xtrain, ytrain)
ypred = model.predict( Xtest)
print('MSE',mean_absolute_error(ytest,ypred))
```
