```python
from sklearn.manifold import MDS
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas
```

```python
df = pandas.read_csv('../Datasets/movies/movies_evaluations.csv')
nmovies,_ = df.values.shape
names = df.values[:,0]
print(nmovies)
df
```

```python
def dist_mov(m1,m2):
    idx = np.logical_and(m1!=0,m2!=0)
    if len(idx) is 0: return 100
    return np.sum(np.abs(m1[idx]-m2[idx]))
diss = np.zeros((nmovies,nmovies))
for i in range(nmovies):
    for j in range(0,i):
        mi = df.values[i,1:]
        mj = df.values[j,1:]
        diss[i,j] = diss[j,i] = dist_mov(mi,mj)
print(diss)
    
```

```python
model = MDS(dissimilarity='precomputed')
X2 = model.fit_transform(diss)
```

```python
plt.figure()
plt.plot(X2[:,0],X2[:,1],'.',color='lightgray')
for i in range(nmovies):
    plt.text(X2[i,0],X2[i,1],names[i][:10])
plt.show()
```
